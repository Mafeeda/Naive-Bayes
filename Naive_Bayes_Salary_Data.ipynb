## Data Preprocessing

```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
number = LabelEncoder()
for i in string_columns:
       train_data[i]= number.fit_transform(train_data[i])
       test_data[i] = number.fit_transform(test_data[i])
```

```python
train_data
```

```python
test_data
```

```python
col_names = train_data.columns
col_names
```

```python
len(col_names)
```

```python
train_data
```

```python
x_train = train_data[col_names[0:13]].values
y_train = train_data[col_names[13]].values
x_test  = test_data[col_names[0:13]].values
y_test  = test_data[col_names[13]].values
```

```python
# Normalmization

def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
x_train = norm_func(x_train)
x_test =  norm_func(x_test)
```

## Naive Bayes Model

```python
M_model = MB()

train_pred_multi = M_model.fit(x_train,y_train).predict(x_train)
test_pred_multi  = M_model.fit(x_train,y_train).predict(x_test)
```

```python
# Train Accuracy

train_acc_multi =  np.mean(train_pred_multi==y_train)
train_acc_multi
```

```python
# Test Accuracy

test_acc_multi = np.mean(test_pred_multi==y_test)
test_acc_multi 
```

```python
# Confusion Matrix

confusion_matrix = confusion_matrix(y_test, test_pred_multi)
confusion_matrix

```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,test_pred_multi))
```

## Gaussian Naive Bayes - 

```python
G_model = GB()
train_pred_gau = G_model.fit(x_train,y_train).predict(x_train)
test_pred_gau  = G_model.fit(x_train,y_train).predict(x_test)



```

```python
# Train Accuracy
train_acc_gau = np.mean(train_pred_gau==y_train)
train_acc_gau 
```

```python
# Test Accuracy
test_acc_gau = np.mean(test_pred_gau==y_test)
test_acc_gau
```

```python
## Confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, test_pred_gau)
confusion_matrix
```

```python
# Accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,test_pred_gau))
```
